```python
# importing required packages
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 

# define matplotlib style
plt.style.use("ggplot")

def round_to_ndigits(i,n):
    return float('{:g}'.format(float('{:.{p}g}'.format(i, p=n))))

round_num = np.vectorize(lambda x: round_to_ndigits(x, 3))#, otypes=np.float)

```

```python
# headers for each column
roundtripTime_header = "Roundtrip Time (s)"
payloadSize_header = "Payload Size (bytes)"


# importing data 
raw_data = pd.read_csv('data.csv')
raw_data.sort_values(by=[roundtripTime_header, payloadSize_header], ascending=True, inplace=True)
data = pd.DataFrame(columns=raw_data.columns)
payloadSizes = raw_data[payloadSize_header].unique()

for payload in payloadSizes:
    payloadData = raw_data[(raw_data[payloadSize_header] == payload)]
    data = data.append(payloadData.mean(), ignore_index=True)

data = data.sort_values(by=[payloadSize_header], ascending=True).reset_index(drop=True)

print(data)


# importing data 
raw_data = pd.read_csv('data1.csv')
raw_data.sort_values(by=[roundtripTime_header, payloadSize_header], ascending=True, inplace=True)
data1 = pd.DataFrame(columns=raw_data.columns)
payloadSizes = raw_data[payloadSize_header].unique()

for payload in payloadSizes:
    payloadData = raw_data[(raw_data[payloadSize_header] == payload)]
    data1 = data1.append(payloadData.mean(), ignore_index=True)

data1 = data1.sort_values(by=[payloadSize_header], ascending=True).reset_index(drop=True)

print(data1)



# importing data 
raw_data = pd.read_csv('data2.csv')
raw_data.sort_values(by=[roundtripTime_header, payloadSize_header], ascending=True, inplace=True)
data2 = pd.DataFrame(columns=raw_data.columns)
payloadSizes = raw_data[payloadSize_header].unique()

for payload in payloadSizes:
    payloadData = raw_data[(raw_data[payloadSize_header] == payload)]
    data2 = data2.append(payloadData.mean(), ignore_index=True)

data2 = data2.sort_values(by=[payloadSize_header], ascending=True).reset_index(drop=True)


print(data2)


# importing data 
raw_data = pd.read_csv('data3.csv')
raw_data.sort_values(by=[roundtripTime_header, payloadSize_header], ascending=True, inplace=True)
data3 = pd.DataFrame(columns=raw_data.columns)
payloadSizes = raw_data[payloadSize_header].unique()

for payload in payloadSizes:
    payloadData = raw_data[(raw_data[payloadSize_header] == payload)]
    data3 = data3.append(payloadData.mean(), ignore_index=True)

data3 = data3.sort_values(by=[payloadSize_header], ascending=True).reset_index(drop=True)


print(data3)


# importing data 
raw_data = pd.read_csv('1_data3.csv')
raw_data.sort_values(by=[roundtripTime_header, payloadSize_header], ascending=True, inplace=True)
data3_1 = pd.DataFrame(columns=raw_data.columns)
payloadSizes = raw_data[payloadSize_header].unique()

for payload in payloadSizes:
    payloadData = raw_data[(raw_data[payloadSize_header] == payload)]
    data3_1 = data3_1.append(payloadData.mean(), ignore_index=True)

data3_1 = data3_1.sort_values(by=[payloadSize_header], ascending=True).reset_index(drop=True)


print(data3_1)

raw_data = pd.read_csv('1_data2.csv')
raw_data.sort_values(by=[roundtripTime_header, payloadSize_header], ascending=True, inplace=True)
data2_1 = pd.DataFrame(columns=raw_data.columns)
payloadSizes = raw_data[payloadSize_header].unique()

for payload in payloadSizes:
    payloadData = raw_data[(raw_data[payloadSize_header] == payload)]
    data2_1 = data2_1.append(payloadData.mean(), ignore_index=True)

data2_1 = data2_1.sort_values(by=[payloadSize_header], ascending=True).reset_index(drop=True)


print(data2_1)


raw_data = pd.read_csv('1_data1.csv')
raw_data.sort_values(by=[roundtripTime_header, payloadSize_header], ascending=True, inplace=True)
data1_1 = pd.DataFrame(columns=raw_data.columns)
payloadSizes = raw_data[payloadSize_header].unique()

for payload in payloadSizes:
    payloadData = raw_data[(raw_data[payloadSize_header] == payload)]
    data1_1 = data1_1.append(payloadData.mean(), ignore_index=True)

data1_1 = data1_1.sort_values(by=[payloadSize_header], ascending=True).reset_index(drop=True)


print(data1_1)


raw_data = pd.read_csv('1_data.csv')
raw_data.sort_values(by=[roundtripTime_header, payloadSize_header], ascending=True, inplace=True)
data_1 = pd.DataFrame(columns=raw_data.columns)
payloadSizes = raw_data[payloadSize_header].unique()

for payload in payloadSizes:
    payloadData = raw_data[(raw_data[payloadSize_header] == payload)]
    data_1 = data_1.append(payloadData.mean(), ignore_index=True)

data_1 = data_1.sort_values(by=[payloadSize_header], ascending=True).reset_index(drop=True)


print(data_1)

```
